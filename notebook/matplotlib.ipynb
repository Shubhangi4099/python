line plot

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x= np.arange(1,11)
x
```

```python
y=x*2
y
```

```python
plt.plot(x,y)
plt.show()
```

addding title and lables

```python
plt.plot(x,y)
plt.title("line plot")
plt.xlabel("x-lable")
plt.ylabel("y-lable")
plt.show()
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth =3)
plt.title("line plot")
plt.xlabel("x-lable")
plt.ylabel("y-lable")
plt.show()
```

adding two line in the same plot

```python
x=np.arange(1,11)
y1=2*x
y2=5*x
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=3)
plt.plot(x,y2,color='r',linestyle='-',linewidth=5)
plt.grid(True)
plt.title("line plot")
plt.xlabel("x-lable")
plt.ylabel("y-lable")
plt.show()
```

adding sub plots

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='blue',linestyle=':',linewidth=3)

plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle='-',linewidth=4)
plt.show()
```

bar plot

```python
student={'bob':56,'sam':66,'romeo':30}
```

```python
names=list(student.keys())
values=list(student.values())
values
```

```python
plt.bar(names,values,color='r')
plt.show()
```

```python
plt.bar(names,values)
plt.grid(True)
plt.title("bar plot")
plt.xlabel("names")
plt.ylabel("values")
plt.show()
```

horizontal bar plot

```python
plt.barh(names,values,color="r")
plt.grid(True)
plt.title("bar plot")
plt.xlabel("values")
plt.ylabel("names")
plt.show()
```

scatter plot

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,5,9,4,3,2,1,7,6]
b=[5,9,6,5,6,4,6,8,7]
```

```python
plt.scatter(x,a)
plt.grid(True)
plt.show()
```

changinging marks

```python
plt.scatter(x,a,marker='*',c='g',s=100)
plt.grid(True)
plt.show()
```

adding two markers in the same plot 

```python
plt.scatter(x,a,marker='*',c='r',s=100)
plt.scatter(x,b,marker='.',c='g',s=200)
plt.grid(True)
plt.show()
```

adding sub ploting

```python
plt.subplot(1,2,1)
plt.scatter(x,a,marker='*',c='r',s=100)
plt.grid()
plt.subplot(1,2,2)
plt.scatter(x,b,marker='.',c='g',s=200)
plt.grid()
plt.show()
```
